```python
import pandas as pd
```

```python
data=pd.read_csv('gun-violence-data.csv')
data.head()
```

```python
data.info()
```

```python
data=data.iloc[:,:7]
data.head(10)
```

```python
data.drop(columns=["incident_id","city_or_county","address"],inplace= True)
data.head()
```

```python
#import liberary from datetime

from datetime import datetime

#convert date to datetime format

data['date'] =pd.to_datetime(data['date'])

#extract year

data['year'] = data['date'].dt.year

#extract month

data['month'] = data['date'].dt.month

#extract day

data['day'] = data['date'].dt.day

#extract day of week

data['day_of_week'] = data['date'].dt.dayofweek

# sum of number of death and injuries

data['total_victim'] = data['n_killed'] + data['n_injured']

data.head()
```

```python
# REPLACE NUMERIC DAY OF WEEK VALUES WITH TEXT

data['data_of_week'] = data['day_of_week'].replace({0: 'Monday', 1: 'Tuesday', 2: 'Wednesday', 3: 'Thursday', 4: 'Friday', 5: 'Saturday', 6: 'Sunday'})
data.head()
```

```python
data.isnull().sum()
```

```python
# FIND THE YEARS

data.groupby('year')['total_victim'].sum()
```

```python
# FIND THE MONTHS

data.groupby('month')['total_victim'].count()
```

```python
# FIND THE DAY

data.groupby('day')['total_victim'].value_counts()
```

```python
# FIND THE DAY OF WEEK

data.groupby('day_of_week')['total_victim'].value_counts()
```

```python
# FIND THE STATE

data.groupby('state')['total_victim'].count()
```
