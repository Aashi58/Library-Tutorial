```python
# arange()         # array of range
# Deep copy

ls=[12,24,35,36,45]
ls1=ls.copy()
print(ls)
print(ls1)
ls1[2]="Aashish"
print(ls)
print(ls1)
```

```python
#numpy linspace
# this function is similar to arange()function. in this function instead of step the number of evenly spaced values between the interval is specified.

import numpy as np
x=np.linspace(10,20,5)
print(x)
```

```python
# endpoint set to false

x=np.linspace(10,20,5, endpoint=False)
print(x)
```

```python
# find the restep value

x=np.linspace(1,2,5, retstep=True)
print(x)
```

```python
#numpy hstack() function
# horizontally stack numpy arrays using numpy.hstack()

x=np.array((3,5,7))
print(x)
print()
y=np.array((6,8,9))
print(y)
print()
np.hstack((x,y))
```

```python
#numpy vstack

import numpy as np
x=np.array([3,5,7])
print(x)
print()
y=np.array([5,7,9])
print(y)
print()
np.vstack((x,y))
```
